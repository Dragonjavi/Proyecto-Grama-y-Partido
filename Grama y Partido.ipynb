### Conteo de tallos de Grama

```python
ImagenContornos = None

def LeerImagen(l):
    Imagen = cv2.imread(l)
    return Imagen, Imagen.copy()

# Tranformacion de la imagen a grises
def PreprocessImage(i):
    return cv2.GaussianBlur(cv2.equalizeHist(cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)), (5, 5), 0)
# Deteccion de contornos usando canny 
def DetectarContornos(i, min, max):
    ImagenBordes = cv2.Canny(i, min, max)
    contornos, _ = cv2.findContours(ImagenBordes.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    return ImagenBordes, contornos

# Definimos contorno de hoja en un area
def ContarHojasDeGrama(contornos):
    total_hojas = 0
    for cn in contornos:
        area = cv2.contourArea(cn)
        # Filtrar contornos pequeos (ajusta el valor segn sea necesario)
        if area > 100:
            total_hojas += 1
    return total_hojas

# Mostramos el  contorno
def DibujarContornos(c):
    global ImagenContornos
    i = 0
    for cn in c:    
        cv2.drawContours(ImagenContornos, cn, -1, (57, 255, 20), 2)    
        x, y, w, h = cv2.boundingRect(cn)  
        cv2.putText(ImagenContornos, str(i+1), (x, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (57, 255, 20), lineType=cv2.LINE_AA)     
        i += 1
    return i

def Plotear(t, im, s, n):
    for i in range(n):
        plt.subplot(2, 2, i+1)
        plt.title(t[i])
        if s[i]:
            plt.imshow(im[i], cmap='gray', vmin=0, vmax=255)        
        else:
            plt.imshow(cv2.cvtColor(im[i], cv2.COLOR_BGR2RGB))        
        plt.xticks([])
        plt.yticks([])
    plt.show()  

# lectura de todas las funciones
def ContarHojasDeGramaEnJardin(location):      
    global ImagenContornos
    Imagen, ImagenContornos = LeerImagen(location)
    ImagenGauss = PreprocessImage(Imagen)
    ImagenBordes, contornos = DetectarContornos(ImagenGauss, 50, 300)
    total_hojas = ContarHojasDeGrama(contornos)
    total_tallos = total_hojas // 5  # Estimacin de tallos dividiendo el total de hojas entre 5
    Titulos = ['Original', 'Gaussiana', 'Bordes de la imagen', f'Estimacin de tallos: {total_tallos}']
    Imagenes = [Imagen, ImagenGauss, ImagenBordes, ImagenContornos]
    Types = [0, 1, 0, 0]
    Plotear(Titulos, Imagenes, Types, 4)

def main():    
    ContarHojasDeGramaEnJardin('Jardin 30cm x 1m.jpg')

if __name__ == '__main__':
    main()
```

```python
import nltk
import spacy
import re
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Descarga los recursos necesarios de NLTK (puede requerir descargar ms recursos segn las necesidades).
nltk.download('punkt')
nltk.download('stopwords')

# Carga el modelo de spaCy en espaol.
nlp = spacy.load('es_core_news_sm')

# Cargar la narracin desde el archivo
with open("Final copa del mundo.txt", "r", encoding="utf-8") as file:
    narracion = file.read()

# Limpieza de texto (resto del cdigo permanece igual)
narracion = re.sub(r'[^a-zA-Z\s]', '', narracion)  
narracion = narracion.lower()  
sentences = nltk.sent_tokenize(narracion)  

# Inicializa los contadores para Argentina y Francia
argentina_stats = {
    "goles": 0,
    "tiros de esquina": 0,
    "faltas": 0,
    "tarjetas amarillas": 0
}

francia_stats = {
    "goles": 0,
    "tiros de esquina": 0,
    "faltas": 0,
    "tarjetas amarillas": 0
}

# Procesamiento de las oraciones
current_team = None  

for sentence in sentences:
    doc = nlp(sentence)

    for token in doc:
        if token.text == "argentina":
            current_team = "argentina"
        elif token.text == "francia":
            current_team = "francia"

        if current_team == "argentina":
            if token.text == "gol":
                argentina_stats["goles"] += 1
            elif token.text == "tiro" and "esquina" in [t.text for t in token.head.children]:
                argentina_stats["tiros de esquina"] += 1
            elif token.text == "falta":
                argentina_stats["faltas"] += 1
            elif token.text == "amarilla":
                argentina_stats["tarjetas amarillas"] += 1

        elif current_team == "francia":
            if token.text == "gol":
                francia_stats["goles"] += 1
            elif token.text == "tiro" and "esquina" in [t.text for t in token.head.children]:
                francia_stats["tiros de esquina"] += 1
            elif token.text == "falta":
                francia_stats["faltas"] += 1
            elif token.text == "amarilla":
                francia_stats["tarjetas amarillas"] += 1

# Wordcloud de la narracin
wordcloud = WordCloud(width=800, height=400).generate(narracion)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# Resultados
print("Estadsticas del partido para Argentina:")
for stat, value in argentina_stats.items():
    print(f"{stat.capitalize()}: {value}")

print("\nEstadsticas del partido para Francia:")
for stat, value in francia_stats.items():
    print(f"{stat.capitalize()}: {value}")
```
